```python
type(cereals)
```

```python
cereals.drop(['shelf','rating'],axis=1,inplace=True)
cereals.set_index("name",inplace=True)
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
imp.fit(cereals)
#imputed_cereals = pd.DataFrame(imputer.transform(cereals),columns=cerealsals.columns)
```

```python
imputed_cereals = pd.DataFrame(imp.fit_transform(cereals),columns=cereals.columns)
```

```python
type(imputed_cereals)
```

```python
imputed_cereals.head(5)
```

```python
scaler=StandardScaler()
scaler.fit(imputed_cereals)
scaled_cereals=pd.DataFrame(scaler.transform(imputed_cereals),columns=cereals.columns)
scaled_cereals
```

```python
type(scaled_cereals)
```

```python
kmeans_cereals=KMeans(n_clusters=5, random_state=123)
kmeans_cereals.fit(scaled_cereals)
```

```python
kmeans_cereals.cluster_centers_
```

```python
kmeans_cereals.inertia_
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
wss= {}
for k in range(2, 21):
    kmeans_loop = KMeans(n_clusters=k,n_init=30,n_jobs=2,random_state=1000,verbose=0).fit(scaled_cereals)
    clusters = kmeans_loop.labels_
    labels = kmeans_loop.predict(scaled_cereals)
    print(silhouette_score(scaled_cereals, labels))
    wss[k] = kmeans_loop.inertia_ 
```

```python
import matplotlib.pyplot as plt
plt.figure()
# %matplotlib notebook 
plt.plot(list(wss.keys()),list(wss.values()) ,marker='o')
plt.grid()
plt.xlabel('Number of clusters')
plt.ylabel('Total within sum of squares')
```

```python
best_kmeans = KMeans(n_clusters=8, random_state=1240)
best_kmeans.fit(scaled_cereals)
best_kmeans_labels = best_kmeans.predict(scaled_cereals)
```

```python
cereals = pd.read_csv("Cereals.csv")
cereals['label'] = cereals['name']+ ' (' + cereals['shelf'].astype(str) + " - " + round(cereals['rating'],2).astype(str) + ')'
cereals.drop(['name','shelf','rating'], axis=1, inplace=True)
```

```python
cereal_label=cereals['label']
cereals.drop('label', axis=1, inplace=True)
```

```python
kmeans_results = pd.DataFrame({"label":cereal_label,"kmeans_cluster":best_kmeans_labels}) 
kmeans_results.head() 
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
linkage_matrix = linkage(scaled_cereals,method='ward',metric='euclidean')
```

```python
import matplotlib.pyplot as plt
dendrogram(linkage_matrix,labels=cereal_label.as_matrix())
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
## Instantiating object
agg_clust = AgglomerativeClustering(n_clusters=6, affinity='euclidean', linkage='ward')
## Training model and return class labels
agg_clusters = agg_clust.fit_predict(scaled_cereals)
## Label - Cluster
agg_result = pd.DataFrame({"label":cereal_label,"agg_cluster":agg_clusters}).sort_values('agg_cluster')
agg_result.head()
```
